## Performing EDA (Exploratory data analysis ): (Univariate Analysis)

## (1) Categorical Data:

```python
#sns.countplot(df['Ad Supported'])
```

```python
# Create a count plot
sns.countplot(data=df, x='Ad Supported')

# Add labels and a title
plt.xlabel('Ad Supported')
plt.ylabel('Count')
plt.title('Distribution of Ad Supported')

# Show the plot
plt.show()
```

```python
print(df['Ad Supported'].value_counts())
df['Ad Supported'].value_counts().plot(kind = 'bar')
```

```python
# Pie chart
df['Ad Supported'].value_counts().plot(kind = 'pie', autopct='%.2f')
```

## (2) Numerical Data:

```python
#Histogram
plt.hist(df['Rating'], bins = 5)
```

```python
#Displot
sns.distplot(df['Rating'])
```

```python
#Displot
sns.displot(df['Rating'])
```

```python
#Boxplot
sns.boxplot(df['Rating'])
```

```python
print(df['Rating'].min())
print(df['Rating'].max())
print(df['Rating'].mean())
```

```python
df['Rating'].skew()
```
