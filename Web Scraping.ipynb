```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#-----------------------------------------------------LAHORE-----------------------------------------------------------
```

```python
url = "http://lhr.nu.edu.pk/faculty/"
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
else:
    print("Failed to retrieve the web page. Status code:", response.status_code)
```

```python
faculty = soup.findAll('div', class_ = 'facultyCard')
```

```python
id_=[]
name=[]
email=[]
desig=[]
HEC=[]
Education=[]
Extension=[]
img=[]
profile=[]
Department=[]
```

```python
def extract_education(profile_url):
    
    total=[]
    
    response = requests.get(profile_url)
    profile_soup = BeautifulSoup(response.content, "html.parser")

    ul_element = profile_soup.find('div', class_='text-justify').find('ul')
    first_li_tag = ul_element.find('li')
    if first_li_tag:
        education_text = first_li_tag.text
        total.append(education_text)
    else:
        total.append('')
    total.append(profile_soup.find('span', class_ = 'small').text.split(':')[-1])
    return total
```

```python
for i in range (len(faculty)):
    fac = faculty[i]
    try:
        id_.append(int(fac.find('a', class_ = 'faculty-link')['href'].split('/')[-1]))
    except:
        id_.append(-1)
    name.append(fac.find('h5', class_ = 'text-center').text)
    email.append(fac.find('p', class_ = 'mb-0').text)
    desig.append(fac.find('p', class_ = 'small').text.strip().split('\n')[0])
    result = fac.find('p', class_='small').text.strip().split('\n')
    if any('HEC APPROVED PHD SUPERVISOR' in text.upper() for text in result):
        HEC.append(True)
    else:
        HEC.append(False)
    img.append('https://lhr.nu.edu.pk' + fac.find('img', class_ = 'card-img-top')['src'])
    profile_url = 'https://lhr.nu.edu.pk' + fac.find('a', class_ = 'faculty-link')['href']
    profile.append(profile_url)
    edu = extract_education(profile_url)
    Education.append(edu[0].strip())
    Extension.append(edu[1].strip())
    Department.append(fac.find('a',class_='faculty-link')['href'].split('/')[1])
       
```

```python
id_count = len(id_)
name_count = len(name)
email_count = len(email)
desig_count = len(desig)
HEC_count = len(HEC)
Education_count = len(Education)
Extension_count = len(Extension)
img_count = len(img)
profile_count = len(profile)
Department_count = len(Department)

print("id_ count:", id_count)
print("name count:", name_count)
print("email count:", email_count)
print("desig count:", desig_count)
print("HEC count:", HEC_count)
print("Education count:", Education_count)
print("Extension count:", Extension_count)
print("img count:", img_count)
print("profile count:", profile_count)
print("Department:", Department_count)
```

```python
pd.set_option('display.max_colwidth', None)

df_lhr = pd.DataFrame({
    'ID': id_,
    'Name':name,
    'Designation':desig,
    'HEC Approved PHD Supervisor': HEC,
    'Highest Education' : Education,
    'Email':email,
    'Department':Department,
    'Extension':Extension,
    'ImageURL':img,
})
```

```python
df_lhr.to_csv('LAHORE_FACULTY.csv', index=False) #end
```

```python
for j in range(len(url)):
    response = requests.get(url[j])
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
    else:
        print("Failed to retrieve the web page. Status code:", response.status_code)
    faculty = soup.findAll('div', class_ = 'gdlr-core-personnel-list-image gdlr-core-media-image gdlr-core-hover-element gdlr-core-zoom-on-hover')
    for i in range(len(faculty)):
        profile = faculty[i].find('a')['href']
        r2 = requests.get(profile)
        soup2 = BeautifulSoup(r2.content, 'html.parser')
        try:
            id_.append (int(soup2.find('script', id='burst-js-extra').text.split(',')[1].split(':')[1][1:-1]))
        except:
            id_.append('N/A')
        name.append(soup2.find('h3', class_='gdlr-core-title-item-title').text.split(',')[0])
        desig.append(soup2.find('span', class_='gdlr-core-title-item-caption gdlr-core-info-font gdlr-core-skin-caption').text.split('\n')[0])
        try:
            soup2.find('span', class_='gdlr-core-title-item-caption gdlr-core-info-font gdlr-core-skin-caption').text.split('\n')[-1]
            HEC.append(True)
        except:
            HEC.append(False)
        try:
            Education.append(soup2.find('span', class_='gdlr-core-icon-list-content').text)
        except:
            Education.append('N/A')
        try:
            email.append(soup2.find('div', class_='kingster-type-email').text)
        except:
            email.append('N/A') 
        
        #Department.append(soup2.find('h1'), class_='kingster-page-title'.text)
        Department.append(total_department[j])
        try:
            Extension.append(soup2.find('div', class_='kingster-type-phone').text)
        except:
            Extension.append('N/A') 
        try:
            img.append(soup2.find('a', class_='gdlr-core-lightgallery')['href'])
        except:
            img.append('N/A') 
        
```

```python
id_count = len(id_)
name_count = len(name)
email_count = len(email)
desig_count = len(desig)
HEC_count = len(HEC)
Education_count = len(Education)
Extension_count = len(Extension)
img_count = len(img)
Department_count = len(Department)

print("id_ count:", id_count)
print("name count:", name_count)
print("email count:", email_count)
print("desig count:", desig_count)
print("HEC count:", HEC_count)
print("Education count:", Education_count)
print("Extension count:", Extension_count)
print("img count:", img_count)
print("Department:", Department_count)
```

```python
file_path = r'C:\Users\HP1\Desktop\Jupyter_Notebook\FAST NUCES Islamabad.html'
html_file = open(file_path, 'r')
html_content = html_file.read()
```

```python
facultytest = BeautifulSoup(html_content, 'html.parser')
facultytest.prettify()
```

```python
faculty = facultytest.findAll('div', class_ = 'our-team')
```

```python
id_=[]
name=[]
email=[]
desig=[]
HEC=[]
Education=[]
Extension=[]
img=[]
Department=[]
```

```python
def get_education(urls): 
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC

    options = webdriver.ChromeOptions()
    options.add_argument("--headless") 
    driver = webdriver.Chrome(options=options)

    url = urls
    driver.get(url)

    wait = WebDriverWait(driver, 10)
    ajax_element = wait.until(EC.presence_of_element_located((By.ID, "DegreeName"))) 
    ajax_content = ajax_element.text
   
    lines = ajax_content.strip().split('\n')
    first_line = lines[0]
  
    driver.quit()
    return first_line
```

```python
def Get_Image_Url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    imgurl = "http://isb.nu.edu.pk" + str(soup.find('div',{'class':'img-thumbnail'}).find('img')['src'])
    
    hec_edu = soup.findAll('label')
    flag = True
    
    for label in hec_edu:
        if label.text.strip().upper() == 'HEC APPROVED PHD SUPERVISOR':
            HEC.append(True)
            flag = True
            break;
        else:
            flag = False
    if flag == False:
        HEC.append(False)
    
    Extension.append(soup.findAll('span')[4].text.split(':')[-1].strip())
    
    Department.append(soup.find('li',class_='breadcrumb-item active').text.strip())
    
    return imgurl 
```

```python
for i in range (len(faculty)):
    fac = faculty[i]
    try:
        id_.append(int(fac.find('a')['href'].split('/')[-1]))
    except:
        id_.append(-1)        
    name.append(fac.find('h3', class_ = 'title').text.strip())
    email.append(fac.find('p').text.strip())
    desig.append(fac.find('span', class_ = 'post').text.strip().split('\n')[0])
    imgurl = fac.find('a')['href']
    img.append(Get_Image_Url(imgurl)) 
    edu = get_education(imgurl)
    Education.append(edu)
    #degree_span = soup.find('span', id='DegreeName')
    #labels = degree_span.find('label')
    #if labels:
```

```python
len(faculty)
id_count = len(id_)
name_count = len(name)
email_count = len(email)
desig_count = len(desig)
HEC_count = len(HEC)
Education_count = len(Education)
Extension_count = len(Extension)
img_count = len(img)
Department_count = len(Department)

print("id_ count:", id_count)
print("name count:", name_count)
print("email count:", email_count)
print("desig count:", desig_count)
print("HEC count:", HEC_count)
print("Education count:", Education_count)
print("Extension count:", Extension_count)
print("img count:", img_count)
print("Department:", Department_count)
```
